# IQR


```python
ranges = pd.DataFrame(columns=['names', 'lower_bound', 'upper_bound'])
```

```python
def detect_anomalies_iqr(data_column,name):
    Q1 = data_column.quantile(0.25)
    Q3 = data_column.quantile(0.75)
    IQR = Q3 - Q1
    print(name)
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(lower_bound)
    print(upper_bound)
    print('--------------------------------------------------------------')
    anomalies = (data_column < lower_bound) | (data_column > upper_bound)
    return lower_bound,upper_bound,anomalies
```

```python
dataframe=data
dataframe=dataframe.drop(['DateTime','FertCode'],axis=1)
cols=dataframe.columns
```

```python
anomalies_dict = {}
for column in cols:
    lower_bound, upper_bound, anomalies = detect_anomalies_iqr(dataframe[column], column)
    anomalies_dict[column] = anomalies
    ranges = pd.concat([ranges,pd.DataFrame({'names': [column], 'lower_bound': [lower_bound], 'upper_bound': [upper_bound]})], ignore_index=True)

for column in cols:
    dataframe[f'{column}_Anomaly'] = anomalies_dict[column].astype(int)
```

```python
 columns=['PrVent1', 'RelPrVent1', 'VentGap1', 'PrVent2', 'RelPrVent2', 'VentGap2',
    'PrVent3', 'RelPrVent3', 'VentGap3', 'PrVent4', 'RelPrVent4', 'VentGap4',
    'PrVent5', 'RelPrVent5', 'VentGap5', 'PrVent6', 'RelPrVent6', 'VentGap6',
    'PrVent7', 'RelPrVent7', 'VentGap7', 'PrVent8', 'RelPrVent8', 'VentGap8',
    'PrVent9', 'RelPrVent9', 'VentGap9', 'VentGap10', 'PrVent10', 'RelPrVent10',
    'PrVent11', 'RelPrVent11', 'VentGap11', 'PrVent12', 'RelPrVent12', 'VentGap12']
for i in columns:
  count_ones = dataframe[f'{i}_Anomaly'].eq(1).sum()
  print(f'{i}_Anomaly',' :  ' ,count_ones)
  percent=(count_ones/125)*100
  print('Percentage of Anomalies : ',percent,'%')
  print('-----------------------------------------')
```

```python
import matplotlib.pyplot as plt
colus=['PrVent1','PrVent2','PrVent3','PrVent4','PrVent5','PrVent6','PrVent7','PrVent8','PrVent9','PrVent10','PrVent11','PrVent12']
for i in colus:
  column_name = i
  anomaly_indices = dataframe[dataframe[f'{column_name}_Anomaly'] == 1].index
  plt.figure(figsize=(20, 4))
  plt.plot(dataframe.index, dataframe[column_name], label='Data')

  plt.scatter(anomaly_indices,dataframe.loc[anomaly_indices, column_name], color='red', label='Anomalies')
  lower_bound = ranges[ranges['names']==column_name]['lower_bound']
  upper_bound = ranges[ranges['names']==column_name]['upper_bound']
  recipe_min =150
  recipe_max =160
  plt.axhline(y=lower_bound.iloc[0], color='blue', linestyle='--', label='Lower Bound (IQR)')
  plt.axhline(y=upper_bound.iloc[0], color='blue', linestyle='--', label='Upper Bound (IQR)')
  plt.axhline(y=recipe_min, color='red', linestyle='--', label='Recipe file VentPre1 (min)')
  plt.axhline(y=recipe_max, color='red', linestyle='--', label='Recipe file VentPre2 (max)')


  plt.title(f'Anomalies Detected in {column_name}')
  plt.xlabel('Index')
  plt.ylabel(column_name)
  plt.legend()
  plt.show()
```

# Confidence level

```python
def detect_anomalies_Confidence(data_column,name):
  mean = np.mean(data_column)
  std_dev = np.std(data_column)
  lower_limit = mean - 3 * std_dev
  upper_limit = mean + 3 * std_dev
  print(name)
  print(lower_limit)
  print(upper_limit)
  print('--------------------------------------------------------------')
  outliers = (data_column < lower_limit) | (data_column > upper_limit)
  return outliers
```

```python
anomalies_dict = {}
for column in cols:
    anomalies_dict[column] = detect_anomalies_Confidence(dataframe[column],column)

for column in cols:
    dataframe[f'{column}_Outlier'] = anomalies_dict[column].astype(int)
```

```python
for i in columns:
  print(f'{i}_Outliers',' : ')
  count_ones = dataframe[f'{i}_Outlier'].eq(1).sum()
  print(count_ones)
  percent=(count_ones/125)*100
  print('Percentage of Outliers : ',percent,'%')
  print('-----------------------------------------')
```

```python
result['Date_Hour'] = pd.to_datetime(result['Date']) + pd.to_timedelta(result['Hour'], unit='h')
```

```python
import matplotlib.dates as mdates
plt.figure(figsize=(15, 6))
plt.plot(result['Date_Hour'],result['PrVent1'], marker='o', linestyle='-')
plt.xlabel('Date_Hour')
plt.ylabel('PrVent1')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%y-%m-%d %H:%M'))
plt.xticks(result['Date_Hour'],rotation=90)
plt.tight_layout()
plt.show()
```
